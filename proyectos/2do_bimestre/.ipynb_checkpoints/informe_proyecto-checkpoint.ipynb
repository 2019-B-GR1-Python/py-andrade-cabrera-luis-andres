# Informe del proyecto segundo bimestre

## Introducción


El objetivo de este proyecto consiste en realizar gráficos que relacionen la información obtenida mediante scrapy desde la página web https://www.manganelo.com con la finalidad de analizarlos. 
La información obtenida contiene la información pública referente a los mangas que se ofrecen dentro del sitio web Manganelo hasta la fecha de 18 de febrero de 2019. 

## Librerías utilizadas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plot
import seaborn as sns
import sqlite3
from datetime import datetime
```

## Extracción de datos

```python
"""Obtenido de: https://manganelo.com/genre-all/* """
database = sqlite3.connect('mangas.db')

all_mangas = pd.read_sql_query("select titulo, vistas from Manga order by vistas DESC", database)
```

## Número de visitas por manga

```python
plot.figure(figsize = (15,5))
sns.barplot(x = 'titulo', y = 'vistas', 
            data = all_mangas.head(5)).set_title('Views por anime')
```

## Porcentaje de mangas publicados por día

```python
mangas_by_day = pd.read_sql_query("select ultimaActualizacion, count(ultimaActualizacion) as 'NumMangas' from Manga GROUP by (ultimaActualizacion)"
                                  , database)
mangas_by_day["weekday"] = pd.to_datetime(mangas_by_day["ultimaActualizacion"], format = "%b %d,%y").dt.dayofweek
print(type(mangas_by_day))
mangas_by_day.loc[mangas_by_day["weekday"]==0,"weekday"] = "Monday"
mangas_by_day.loc[mangas_by_day["weekday"]==1,"weekday"] = "Tuesday"
mangas_by_day.loc[mangas_by_day["weekday"]==2,"weekday"] = "Wednesday"
mangas_by_day.loc[mangas_by_day["weekday"]==3,"weekday"] = "Thursday"
mangas_by_day.loc[mangas_by_day["weekday"]==4,"weekday"] = "Friday"
mangas_by_day.loc[mangas_by_day["weekday"]==5,"weekday"] = "Saturday"
mangas_by_day.loc[mangas_by_day["weekday"]==6,"weekday"] = "Sunday"
mangas_by_day_grouped_by = mangas_by_day.groupby(["weekday","NumMangas"])
print(mangas_by_day)

plot.figure(figsize = (15, 5))
sns.barplot(x = 'weekday', y = 'NumMangas', data = mangas_by_day.groupby(["weekday","NumMangas"]).sum().sort_values(by=['NumMangas']).reset_index()).set_title('Mangas por dia de la semana')
plot.xticks(rotation = 90)
```

## Relación votos por vistas por manga

```python
votes_by_views_by_anime = pd.read_sql_query("select titulo, vistas, votos from Manga order by vistas DESC",database)
plot.figure(figsize = (15, 5))
votes_by_views_by_anime["relacion"] = votes_by_views_by_anime["votos"] / votes_by_views_by_anime["vistas"]
sns.barplot(x='titulo', y='relacion', data=votes_by_views_by_anime.head(7))
```

## Número de mangas por genero

```python
numero_mangas_genero = pd.read_sql_query("select Genero.nombre, count(GeneroManga.idGenero) as 'NumeroMangas' from Genero, GeneroManga, Manga where Manga.idManga = GeneroManga.idManga and Genero.idGenero = GeneroManga.idGenero group by GeneroManga.idGenero order by (NumeroMangas) DESC", database).head(10)
generos = numero_mangas_genero["nombre"].to_numpy()
numero_mangas = numero_mangas_genero["NumeroMangas"].to_numpy()
plot.figure(figsize = (15, 15)) 
plot.pie(numero_mangas, 
         labels = generos, 
         autopct = "%1.3f%%")
plot.title('Número de mangas publicados por genero')
```

## Porcentaje de mangas publicados por autor

```python
numero_mangas_autor = pd.read_sql_query("select Autor.nombre, count(AutorManga.idAutor) as 'NumeroMangas' from Autor, AutorManga, Manga where Manga.idManga = AutorManga.idManga and Autor.idAutor = AutorManga.idAutor group by AutorManga.idAutor order by (NumeroMangas) DESC", database).head(10)
autores = numero_mangas_autor["nombre"].to_numpy()
numero_mangas = numero_mangas_autor["NumeroMangas"].to_numpy()
plot.figure(figsize = (15, 15)) 
plot.pie(numero_mangas, 
         labels = autores, 
         autopct = "%1.3f%%")
plot.title('Número de mangas publicados por genero')
```

## Relación genero por rating

```python
genre_rating = pd.read_sql_query("select Genero.nombre, count(GeneroManga.idGenero) as 'NumeroMangas', SUM(Manga.rating)/count(GeneroManga.idGenero) as 'TotalRating' from Genero, GeneroManga, Manga where Manga.idManga = GeneroManga.idManga and Genero.idGenero = GeneroManga.idGenero group by GeneroManga.idGenero order by ('NumeroMangas') DESC"
                                  , database)
plot.figure(figsize = (15, 5))
genre_rating["relacion"] = genre_rating["TotalRating"] / genre_rating["NumeroMangas"]
sns.barplot(x='nombre', y='relacion', data=genre_rating.sort_values(by = "NumeroMangas", ascending = False).head(10))
```

## Número de capitulos por manga

```python
caps_manga = pd.read_sql_query("select capitulos, titulo from Manga order by(capitulos) DESC", database)
plot.figure(figsize = (15, 5))
sns.barplot(x='titulo', y='capitulos', data=caps_manga.head(7))
```

## Número de mangas publicados por mes

```python
mangas_by_day = pd.read_sql_query("select ultimaActualizacion, count(ultimaActualizacion) as 'NumMangas' from Manga GROUP by (ultimaActualizacion) order by(\"NumMangas\") DESC"
                                  , database)
mangas_by_day["month"] = pd.to_datetime(mangas_by_day["ultimaActualizacion"], format = "%b %d,%y").dt.month
print(mangas_by_day)

plot.figure(figsize = (15, 5))
sns.barplot(x = 'month', y = 'NumMangas', data = mangas_by_day.groupby(["month","NumMangas"]).sum().reset_index()).set_title('Mangas por mes')
plot.xticks(rotation = 90)
```

## Porcentaje de mangas por estado

```python
mangas_state = pd.read_sql_query("select estado, count(estado) as \"NumManga\" from Manga group by (estado) ORDER by (\"NumMangas\") DESC", database)
estado = mangas_state["estado"].to_numpy()
numero_mangas = mangas_state["NumManga"].to_numpy()
plot.figure(figsize = (15, 15)) 
plot.pie(numero_mangas, 
         labels = estado, 
         autopct = "%1.3f%%")
plot.title('Número de mangas publicados por estador')
```

## Relación Numero mangas por genero por capitulos

```python
genre_manga_caps = pd.read_sql_query("select Genero.nombre, count(GeneroManga.idGenero) as 'NumeroMangas',SUM(Manga.capitulos)/SUM(Manga.rating) as 'CapitulosGenero' from Genero, GeneroManga, Manga where Manga.idManga = GeneroManga.idManga and Genero.idGenero = GeneroManga.idGenero group by GeneroManga.idGenero order by (\"NumeroMangas\") DESC", database).head(5)
plot.figure(figsize = (15, 5)) 
sns.barplot(x = "nombre", y = "NumeroMangas", hue = "CapitulosGenero", data = genre_manga_caps)
```
